```python
import matplotlib.pyplot as plt
```

```python
edad_x = [20,25,30,35,40,45,50,55,60,65]
salario_cd_y = [2000,3000,3300,4000,4200,4500,4700,6000,7000,8000]
salario_fed_y = [1000,1500,1700,2700,3000,3200,3500,4000,4900,5000]
salario_bed_y = [1500,2000,2500,2700,3500,4000,4500,5000,5500,6000]
```

```python
plt.plot(edad_x,salario_cd_y)
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y)
plt.title("Salario Medio de un Cientifico de Datos")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y, label="Cientifico de datos")
plt.plot(edad_x,salario_fed_y, label="Front End Desarrollador")

plt.title("Salario Medio por edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y, label="Cientifico de datos")
plt.plot(edad_x,salario_fed_y, label="Front End Desarrollador")
plt.plot(edad_x, salario_bed_y, label="Back End Desarrollador")

plt.title("Salario Medio por edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.legend()
plt.show()
```

```python
plt.plot(edad_x,salario_cd_y, label="Cientifico de datos", linewidth=3, marker=".",markersize=15, linestyle="--", color="#d62728")
plt.plot(edad_x,salario_fed_y, label="Front End Desarrollador", linewidth=3)
plt.plot(edad_x, salario_bed_y, label="Back End Desarrollador", linewidth=3)

plt.title("Salario Medio por edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
plt.plot(edad_x,salario_cd_y, label="Cientifico de datos", linewidth=3)
plt.plot(edad_x,salario_fed_y, label="Front End Desarrollador", linewidth=3)
plt.plot(edad_x, salario_bed_y, label="Back End Desarrollador", linewidth=3)

salario_medio = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3

plt.plot(edad_x, salario_medio, label="Salario Medio", linestyle="--", color="#000000", linewidth=3, marker = ".",markersize=15)

plt.title("Salario Medio por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label="Back End Desarrollador", linewidth=2)

salario_medio_y = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3

plt.plot(edad_x, salario_medio, label="Salario Medio", linestyle="--", color="#000000", linewidth=2, marker = ".",markersize=15)

plt.fill_between(edad_x,salario_bed_y,salario_medio_y,
                interpolate =True,alpha=0.33, label="Between") # Alpha: Intensidad del color

plt.title("Salario Medio por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label="Back End Desarrollador", linewidth=2)

salario_medio_y = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3

plt.plot(edad_x, salario_medio, label="Salario Medio", linestyle="--", color="#000000", linewidth=2, marker = ".",markersize=15)

plt.fill_between(edad_x,salario_bed_y,salario_medio_y, where=(salario_bed_y>salario_medio_y),interpolate= True,alpha=0.33, label="Between") # Alpha: Intensidad del color

plt.title("Salario Medio por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.grid()
plt.legend()
plt.show()
```

```python
plt.plot(edad_x, salario_bed_y, label="Back End Desarrollador", linewidth=2)

salario_medio_y = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3

plt.plot(edad_x, salario_medio, label="Salario Medio", linestyle="--", color="#000000", linewidth=2, marker = ".",markersize=15)

plt.fill_between(edad_x,salario_bed_y,salario_medio_y, where=(salario_bed_y>salario_medio_y),interpolate= True,alpha=0.33, label="> Media") # Alpha: Intensidad del color

plt.fill_between(edad_x,salario_bed_y,salario_medio_y, where=(salario_bed_y<salario_medio_y),color="yellow",interpolate= True,alpha=0.33, label="< Media") # Alpha: Intensidad del color

plt.title("Salario Medio por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.grid()
plt.legend()
plt.show()
```

```python
separacion=0.1
edad_np_x=np.arange(len(edad_x))
plt.bar(edad_np_x-2*separacion,salario_cd_y, label="Cientifico de datos",width=separacion)
plt.bar(edad_np_x,salario_fed_y, label="Front End Desarrollador",width=separacion)
plt.bar(edad_np_x+2*separacion, salario_bed_y, label="Back End Desarrollador",width=separacion)

salario_medio_y = np.sum([salario_cd_y,salario_fed_y,salario_bed_y], axis=0)/3

plt.bar(edad_np_x+4*separacion, salario_medio_y, label="Salario Medio", color="#000000",width=2*separacion)

plt.title("Salario Medio por Edad")
plt.xlabel("Edad")
plt.ylabel("Salario Medio")

plt.xticks(ticks=edad_np_x, labels=edad_x)

plt.grid()
plt.legend()
plt.show()
```
